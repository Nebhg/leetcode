## 1. Stack

### The problem it solves

Sometimes the order in which you process things is the reverse of the order you received them. Opening a bracket `(` means nothing until you find its matching `)`. Navigating back in a browser means undoing the most recent page first. Calling a function means the most recently called function must return before the one that called it.

All of these have the same shape: **the last thing in must be the first thing out**. That's a stack.

### The contract (LIFO — Last In, First Out)

| Operation | Requirement | Why |
|-----------|-------------|-----|
| `push(x)` | O(1) — add to top | Must be fast or the structure is useless |
| `pop()` | O(1) — remove from top | Same |
| `peek()` | O(1) — view top without removing | Needed to make decisions without committing |
| `is_empty()` | O(1) | Guards against popping an empty stack |

A Python list already gives us O(1) `append` (push) and `pop` from the end. So the Stack class is just a list with a constrained interface — we only allow operations at one end.

### Why constrain the interface?

Because allowing arbitrary access defeats the purpose. A stack guarantees that the only thing you can ever see is the top — that constraint is what makes problems like bracket matching tractable. If you could randomly access elements, you'd have no structure to reason about.

## 2. Queue

### The problem it solves

A queue processes items in the order they arrive — **first in, first out**. The classic real-world queue: customers at a till, print jobs sent to a printer, HTTP requests arriving at a server. The item that waited longest gets served first.

In algorithms, queues power **BFS (Breadth-First Search)** — exploring a graph level by level. You discover nodes layer by layer because each newly discovered node joins the back of the queue and gets processed in arrival order.

### The contract (FIFO — First In, First Out)

| Operation | Requirement | Why |
|-----------|-------------|-----|
| `enqueue(x)` | O(1) — add to back | — |
| `dequeue()` | O(1) — remove from front | **This is why a plain list fails** |
| `peek()` | O(1) — view front | — |

### Why not just use a list?

A Python list supports O(1) `append` at the back — that's fine for enqueue. But `list.pop(0)` (removing from the front) is **O(n)** because every remaining element must shift left.

Use `collections.deque` instead. It's a doubly-linked list that provides O(1) operations at **both** ends: `append` (right) and `popleft` (left).

## 3. Linked List

### The problem it solves

A list stores items in contiguous memory — which makes indexing O(1) but insertion/deletion in the middle O(n) (because items must shift). A linked list takes the opposite trade: items can live anywhere in memory, connected by pointers.

**The result:**
- Insert or delete at a known position: O(1) — just rewire pointers
- Find an element: O(n) — must follow the chain
- Index access: O(n) — no random access

### The mental model

Think of a linked list as a **treasure hunt**. Each clue (node) tells you where the next clue is. To find clue #5, you must start at clue #1 and follow the chain. You can't jump directly.

### The node

Every linked list is built from nodes. A node holds two things: its own value, and a pointer to the next node. The last node's `next` pointer is `None` — that's how you know you've reached the end.

```
head → [1 | •] → [2 | •] → [3 | None]
```

The list itself only needs to store `head` — the starting point. Everything else is reachable by following `next` pointers.

## 4. Binary Tree

### The problem it solves

Trees model **hierarchical relationships**: file systems, org charts, HTML DOM, mathematical expressions. A binary tree is the simplest variant — each node has at most two children, called `left` and `right`.

### The mental model

Think of a tree as a **set of decisions**. At each node, you go left or right. The path from the root to any node is a sequence of those decisions. This structure allows binary search trees to find elements in O(log n) by halving the search space at each node.

### Traversal orders — and why they matter

The order in which you visit nodes changes what you see:

| Traversal | Order | When to use |
|-----------|-------|-------------|
| **Inorder** | left → node → right | Gets sorted values from a BST |
| **Preorder** | node → left → right | Copies or serialises a tree |
| **Postorder** | left → right → node | Deletes a tree; evaluates expressions |
| **Level order** | row by row | BFS; finding minimum depth |
