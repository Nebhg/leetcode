## 1. Lists

### What is a list?

A list is a **contiguous block of memory** holding an ordered sequence of items. Because items are stored next to each other in memory, Python can jump directly to any element by index in O(1) time — it's just arithmetic: `address = start + (index × item_size)`.

### The mental model

Think of a list like a **numbered row of boxes**. Box 0 is at the start, box -1 is at the end. You can grab any box instantly because you know exactly where it is.

### When does this matter?

- Accessing by index is always O(1) — use this when you need random access
- Appending to the end is O(1) amortised — use this as your default dynamic array
- Inserting or deleting from the middle is O(n) — all subsequent items must shift
- Searching for a value is O(n) — you may need to check every element

### Key operations to know cold

| Operation | Syntax | Notes |
|-----------|--------|-------|
| Access by index | `nums[i]` | Negative index counts from end: `nums[-1]` is last |
| Slice | `nums[start:stop]` | Stop is exclusive. `nums[:3]` = first 3. `nums[2:]` = from index 2 |
| Reverse slice | `nums[::-1]` | Step of -1 = walk backwards |
| Append | `nums.append(x)` | O(1) — always use this to add to end |
| Pop end | `nums.pop()` | O(1) — removes and returns last element |
| Pop index | `nums.pop(i)` | O(n) — shifts elements |
| Sort in place | `nums.sort()` | Modifies original. `sorted(nums)` returns new list |
| Enumerate | `for i, v in enumerate(nums)` | When you need index AND value |
| Comprehension | `[f(x) for x in nums if cond]` | Build new list in one line |

## 2. Dictionaries (Hash Maps)

### What is a dictionary?

A dictionary is a **hash map** — a structure that maps keys to values with O(1) average-case lookup, insert, and delete. Under the hood, Python hashes your key to compute a memory address directly, rather than searching through elements.

### The mental model

Think of a dictionary like a **phonebook with instant lookup**. You don't read through every entry — you open directly to the right page. The key is hashed to tell Python exactly where to look.

### Why does this matter in interviews?

The most common use of a dictionary in interview problems is **trading space for time**. Instead of scanning an array O(n) to find something, you store it in a dict first so you can look it up in O(1) later. This pattern turns many O(n²) brute-force solutions into O(n).

### Three patterns you'll use constantly

**1. Frequency counting** — how often does each item appear?
```python
counts = {}
for item in collection:
    counts[item] = counts.get(item, 0) + 1
```

**2. Complement lookup** — have I seen what I need before?
```python
seen = {}
for i, num in enumerate(nums):
    complement = target - num
    if complement in seen:
        return [seen[complement], i]
    seen[num] = i
```

**3. Grouping** — put items into buckets by some shared property
```python
groups = {}
for item in collection:
    key = derive_key(item)  # e.g. sorted letters for anagram grouping
    if key not in groups:
        groups[key] = []
    groups[key].append(item)
```

### Key operations

| Operation | Syntax | Notes |
|-----------|--------|-------|
| Safe get with default | `d.get(key, default)` | Never raises KeyError |
| Check key exists | `key in d` | O(1) |
| Iterate key-value pairs | `for k, v in d.items()` | Most common loop form |
| Delete key | `del d[key]` | Raises KeyError if missing |
| All keys / values | `d.keys()`, `d.values()` | Return view objects |

````python
# Exercise 2b — Two Sum (complement lookup pattern)
#
# Given a list of numbers and a target, return the indices of the
# two numbers that add up to the target. Each input has exactly one solution.
#
# Input:  nums=[2, 7, 11, 15], target=9
# Output: [0, 1]  (nums[0] + nums[1] = 9)
#
# Think: for each number, what are you looking for?
# How can a dict help you find it in O(1) instead of scanning?
# What do you store as the key? What as the value?

from typing import List

def twoSum(nums: List[int], target: int) -> List[int]:
    seen = {}

    for i, n in enumerate(nums):
        complement = target - n
        if complement in seen:
            return [seen[complement], i]
        seen[n] = i

assert sorted(twoSum([2, 7, 11, 15], 9)) == [0, 1]
assert sorted(twoSum([3, 2, 4], 6)) == [1, 2]
assert sorted(twoSum([3, 3], 6)) == [0, 1]
print('Exercise 2b: passed')
````

## 3. Sets

### What is a set?

A set is an **unordered collection of unique items** with O(1) average lookup. Under the hood it's a hash map where values are discarded — only keys remain.

### The mental model

Think of a set as a **bag that rejects duplicates**. You can check whether something is in the bag instantly, but you can't look things up by position because there's no order.

### When to use a set vs a dict

- Use a **set** when you only care whether something has been *seen* (yes/no)
- Use a **dict** when you also need to remember something *about* what you've seen (e.g. its index, its count)

The classic use case: `seen = set()` to track visited items in O(1) instead of O(n).

### Key operations

| Operation | Syntax | Notes |
|-----------|--------|-------|
| Create from list | `set(my_list)` | Removes duplicates automatically |
| Add | `s.add(x)` | No-op if already present |
| Remove safely | `s.discard(x)` | No error if x not in s |
| Membership | `x in s` | O(1) |
| Intersection | `a & b` | Items in both |
| Union | `a \| b` | Items in either |
| Difference | `a - b` | Items in a but not b |

## 4. Strings

### What makes strings different?

Strings in Python are **immutable** — you cannot change a character in place. Every operation that seems to modify a string actually creates a new one. This has a performance implication: repeatedly concatenating strings in a loop (`s += char`) is O(n²). Use `''.join(list_of_chars)` instead.

### The mental model

Treat a string like a **read-only list of characters**. All slicing and indexing works identically to lists. When you need to build a string incrementally, accumulate characters in a list and join at the end.

### Patterns that come up in interviews

**Checking character types** — `.isalpha()`, `.isdigit()`, `.isalnum()` — used constantly in problems that say "ignore non-alphanumeric characters"

**Two-pointer on string** — treat the string like a list, use left/right indices

**Frequency map of characters** — `counts = {}; for c in s: counts[c] = counts.get(c, 0) + 1`

**Canonical form** — `sorted(s)` or `tuple(sorted(s))` turns a string into a key that is the same for all anagrams

### Key operations

| Operation | Syntax | Notes |
|-----------|--------|-------|
| Lower/upper | `s.lower()`, `s.upper()` | Returns new string |
| Strip whitespace | `s.strip()` | Also `lstrip()`, `rstrip()` |
| Split | `s.split()` | Splits on whitespace by default |
| Join | `sep.join(lst)` | Inverse of split — prefer over concatenation |
| Replace | `s.replace(old, new)` | Returns new string |
| Starts/ends with | `s.startswith(p)`, `s.endswith(p)` | — |
| Char type checks | `c.isalpha()`, `c.isdigit()`, `c.isalnum()` | — |

## 5. Complexity — the thing that ties everything together

Every data structure choice is a complexity trade-off. You need to know these cold because interviewers ask, and because choosing the wrong structure makes your solution wrong at scale.

### The structures at a glance

| Structure | Access | Search | Insert (end) | Insert (middle) | Delete |
|-----------|--------|--------|--------------|-----------------|--------|
| List | O(1) | O(n) | O(1) amortised | O(n) | O(n) |
| Dict | — | O(1) avg | O(1) avg | — | O(1) avg |
| Set | — | O(1) avg | O(1) avg | — | O(1) avg |
| String | O(1) | O(n) | O(n) (immutable) | — | — |

### The key question to ask yourself

> *"What operations do I need to do most often, and what's their cost with each structure?"*

If you're searching repeatedly → don't use a list, use a set or dict.  
If you need to remember something about each item → dict, not set.  
If you only need yes/no membership → set, not dict.
