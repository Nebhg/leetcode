## 1. Lists
The most common data structure. Master every operation cold.

```python
# EXAMPLE — run this cell to see the patterns
nums = [3, 1, 4, 1, 5, 9, 2, 6]

print(nums[0])          # first element -> 3
print(nums[-1])         # last element -> 6
print(nums[-2])         # second to last -> 2
print(nums[1:4])        # slice index 1 up to (not including) 4 -> [1, 4, 1]
print(nums[:3])         # first 3 elements -> [3, 1, 4]
print(nums[5:])         # from index 5 to end -> [9, 2, 6]
print(nums[::-1])       # reversed -> [6, 2, 9, 5, 1, 4, 1, 3]
print(len(nums))        # length -> 8
print(5 in nums)        # membership check -> True
print(nums.index(9))    # find index of value -> 5
print(nums.count(1))    # count occurrences -> 2

nums.append(7)          # add to end
nums.insert(0, 99)      # insert at index
nums.pop()              # remove and return last
nums.pop(0)             # remove and return at index
nums.remove(4)          # remove first occurrence of value
nums.sort()             # sort in place
nums.reverse()          # reverse in place

sorted_copy = sorted(nums)          # sorted returns new list
reversed_copy = list(reversed(nums)) # reversed returns iterator

for i, val in enumerate(nums):      # iterate with index
    pass

squares = [x**2 for x in nums]              # list comprehension
evens = [x for x in nums if x % 2 == 0]    # with filter
```

## 2. Dictionaries (Hashmaps)
Used in ~60% of interview problems. Know these cold.

```python
# EXAMPLE
d = {}                          # empty dict
d = {'a': 1, 'b': 2}            # dict literal

d['c'] = 3                      # add/update key
d['a'] += 1                     # increment value
print(d['a'])                   # access value -> 2
print(d.get('z', 0))            # safe access with default -> 0
print('b' in d)                 # check key exists -> True
del d['b']                      # delete key

for key in d:                   # iterate keys
    pass
for key, val in d.items():      # iterate key-value pairs
    pass
for val in d.values():          # iterate values
    pass

print(list(d.keys()))           # all keys as list
print(list(d.values()))         # all values as list

# THE PATTERN — character/item counting (comes up constantly)
word = 'hello'
counts = {}
for ch in word:
    if ch in counts:
        counts[ch] += 1
    else:
        counts[ch] = 1
# shorthand using get:
counts2 = {}
for ch in word:
    counts2[ch] = counts2.get(ch, 0) + 1
```

## 3. Sets
For deduplication and fast membership checks. O(1) lookup.

```python
# EXAMPLE
s = set()                       # empty set — NOT {} (that's a dict)
s = {1, 2, 3}                   # set literal
s = set([1, 2, 2, 3, 3])        # from list, removes dupes -> {1, 2, 3}

s.add(4)                        # add element
s.discard(99)                   # remove if exists (no error if missing)
s.remove(2)                     # remove (raises error if missing)
print(3 in s)                   # membership check -> True
print(len(s))                   # size

a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
print(a & b)                    # intersection -> {3, 4}
print(a | b)                    # union -> {1, 2, 3, 4, 5, 6}
print(a - b)                    # difference -> {1, 2}

# THE PATTERN — seen set for deduplication
seen = set()
for num in [1, 2, 2, 3, 3, 3]:
    if num in seen:
        print(f'duplicate: {num}')
    seen.add(num)
```

## 5. Two Pointer Pattern
Comes up constantly. Works on sorted arrays and strings.

```python
# EXAMPLE — check if string is palindrome using two pointers
def is_palindrome(s):
    left, right = 0, len(s) - 1
    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    return True

print(is_palindrome('racecar'))  # True
print(is_palindrome('hello'))    # False
```

```python
# Rewrite Two Sum from memory
# Input: nums = [2, 7, 11, 15], target = 9
# Output: [0, 1]
# No looking at previous solution

from typing import List

def twoSum(nums: List[int], target: int) -> List[int]:
    seen = {}
   
    for i, num in enumerate(nums):
       complement = target - num
       if complement in seen:
           return [i, seen[complement]]
       seen[num] = i 

# Test
assert twoSum([2, 7, 11, 15], 9) == [1, 0] or twoSum([2, 7, 11, 15], 9) == [0, 1]
assert twoSum([3, 2, 4], 6) == [1, 2] or twoSum([3, 2, 4], 6) == [2, 1]
print('Two Sum: passed')
```

```python
# Rewrite Valid Anagram from memory
# Input: s = 'anagram', t = 'nagaram'
# Output: True

def isAnagram(s: str, t: str) -> bool:
    s_map = {}
    t_map = {}

    if not s or not t:
        return False
    #count frequency of chars in each string
    # should be equal if its a valid anagram
    for i in s:
        if i in s_map:
            s_map[i] += 1
        else:
            s_map[i] = 1
    
    for j in t:
        if j in t_map:
            t_map[j] += 1
        else:
            t_map[j] = 1

    return s_map == t_map
    
    
# Test
assert isAnagram('anagram', 'nagaram') == True
assert isAnagram('rat', 'car') == False
assert isAnagram('a', 'ab') == False
print('Valid Anagram: passed')
```

```python
# Rewrite Contains Duplicate from memory
# Input: nums = [1,2,3,1]
# Output: True

def containsDuplicate(nums: List[int]) -> bool:
    seen = set()

    for val in nums:
        if val in seen:
            return True
        seen.add(val)
    return False

# Test
assert containsDuplicate([1, 2, 3, 1]) == True
assert containsDuplicate([1, 2, 3, 4]) == False
print('Contains Duplicate: passed')
```
