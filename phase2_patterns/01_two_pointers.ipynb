# Pattern: Two Pointers

## The core insight

Many problems ask you to find a pair of elements in an array that satisfies some condition. The brute-force approach checks every pair — O(n²). Two pointers exploits **structure in the data** (usually sortedness) to do it in O(n).

The idea: instead of one loop inside another, you move two pointers — one from each end, or both from the same end — and each pointer only ever moves forward. The total number of steps is O(n).

## How to recognise this pattern

Ask yourself:
- Is the array sorted, or can I sort it without losing information I need?
- Am I looking for a pair, or comparing elements symmetrically (palindrome)?
- Am I partitioning in-place — keeping some elements, discarding others?

## The two variants

**Opposite ends** — left starts at 0, right at the end, they move toward each other. Use this when searching for pairs in a sorted array, or checking symmetry.

```
left →                    ← right
 [1,  2,  3,  4,  5,  6]
```

**Same direction (read/write)** — both start at 0. One reads every element; the other only advances when it finds something worth keeping. Use this for in-place filtering or deduplication.

```
write →
read  →→→→→→
[1, 1, 2, 3, 3, 4]
```


**Why does sortedness make two pointers work here?**

If `nums[left] + nums[right]` is too large, moving `right` left gives you a smaller value — the only direction that can reduce the sum. If it's too small, moving `left` right is the only thing that increases it. Each move eliminates one element from consideration permanently, so the total work is O(n).

````python
# ============================================================
# SOLUTION
# ============================================================

def twoSumSorted_solution(numbers: List[int], target: int) -> List[int]:
    left, right = 0, len(numbers) - 1
    while left < right:
        total = numbers[left] + numbers[right]
        if total == target:
            return [left + 1, right + 1]
        elif total < target:
            left += 1
        else:
            right -= 1
    return []
````

````python
# ============================================================
# SOLUTION
# ============================================================

def maxArea_solution(height: List[int]) -> int:
    left, right = 0, len(height) - 1
    best = 0
    while left < right:
        area = min(height[left], height[right]) * (right - left)
        best = max(best, area)
        if height[left] < height[right]:
            left += 1   # shorter side limits us — move it, maybe we find something taller
        else:
            right -= 1  # moving the taller can only make things worse
    return best
````


````python
# ============================================================
# SOLUTION
# ============================================================

def threeSum_solution(nums: List[int]) -> List[List[int]]:
    nums.sort()
    result = []
    for i in range(len(nums) - 2):
        if i > 0 and nums[i] == nums[i - 1]:
            continue  # skip duplicate fixed element
        left, right = i + 1, len(nums) - 1
        while left < right:
            total = nums[i] + nums[left] + nums[right]
            if total == 0:
                result.append([nums[i], nums[left], nums[right]])
                while left < right and nums[left] == nums[left + 1]:   left += 1
                while left < right and nums[right] == nums[right - 1]: right -= 1
                left += 1; right -= 1
            elif total < 0:
                left += 1
            else:
                right -= 1
    return result
````
