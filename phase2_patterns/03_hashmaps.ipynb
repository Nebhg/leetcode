# Pattern: Hashmaps

## When to use it
- You need O(1) lookup, insert, or delete
- Counting frequencies of elements
- Grouping elements by some property
- Checking if a complement/pair exists (Two Sum pattern)
- Keywords: 'find pair', 'count occurrences', 'group by', 'anagram'

## Core patterns

### 1. Frequency counter
```python
counts = {}
for item in collection:
    counts[item] = counts.get(item, 0) + 1
```

### 2. Complement lookup (Two Sum)
```python
seen = {}
for i, num in enumerate(nums):
    complement = target - num
    if complement in seen:
        return [seen[complement], i]
    seen[num] = i
```

### 3. Grouping
```python
groups = {}
for item in collection:
    key = get_key(item)  # e.g. sorted letters for anagram grouping
    if key not in groups:
        groups[key] = []
    groups[key].append(item)
```

````python
# EXAMPLE 1 — Group Anagrams (LC #49)
# Group strings that are anagrams of each other
# Input: ['eat','tea','tan','ate','nat','bat']
# Output: [['eat','tea','ate'],['tan','nat'],['bat']]

from typing import List

def groupAnagrams(strs: List[str]) -> List[List[str]]:
    groups = {}
    
    for s in strs:
        key = tuple(sorted(s))  # sorted letters as key — anagrams share the same key
        if key not in groups:
            groups[key] = []
        groups[key].append(s)
    
    return list(groups.values())

print(groupAnagrams(['eat','tea','tan','ate','nat','bat']))
````

````python
# EXAMPLE 2 — Top K Frequent Elements (LC #347)
# Input: [1,1,1,2,2,3], k=2
# Output: [1,2]

def topKFrequent(nums: List[int], k: int) -> List[int]:
    counts = {}
    for num in nums:
        counts[num] = counts.get(num, 0) + 1
    
    # sort by frequency, take top k
    return sorted(counts, key=lambda x: counts[x], reverse=True)[:k]

print(topKFrequent([1,1,1,2,2,3], 2))  # [1, 2]
````

````python
# EXAMPLE 3 — Valid Sudoku style: using sets per row/col/box
# Demonstrates using multiple hashmaps simultaneously

# Checking for duplicates across multiple dimensions at once:
from collections import defaultdict

# defaultdict(set) auto-creates a set when a new key is accessed
# saves you writing: if key not in d: d[key] = set()

rows = defaultdict(set)
cols = defaultdict(set)

data = [(0, 0, 5), (0, 1, 3), (1, 0, 6)]  # (row, col, val)
valid = True

for r, c, val in data:
    if val in rows[r] or val in cols[c]:
        valid = False
        break
    rows[r].add(val)
    cols[c].add(val)

print(valid)
````
