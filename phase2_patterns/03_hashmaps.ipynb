# Pattern: Hash Maps

## The core insight

Hash maps let you **trade memory for time**. Instead of scanning an array O(n) to find something, you spend one O(n) pass to build a map — then every future lookup is O(1). This turns many O(n²) brute-force solutions into O(n).

## How to recognise this pattern

Ask yourself:
- "Have I seen this before?" — you need to remember past elements (seen set or dict)
- "How many times does each X appear?" — frequency counting
- "Can I group things by a shared property?" — grouping by a computed key
- "I need to find a complement / pair" — Two Sum style

## The three core patterns

**1. Frequency counting** — count how often each item appears
```python
counts = {}
for item in collection:
    counts[item] = counts.get(item, 0) + 1
```

**2. Complement lookup (Two Sum)** — store what you've seen, look for what you need
```python
seen = {}  # value → index
for i, num in enumerate(nums):
    complement = target - num
    if complement in seen:
        return [seen[complement], i]
    seen[num] = i
```

**3. Grouping** — bucket items by a shared canonical key
```python
groups = {}  # key → [items]
for item in collection:
    key = derive_key(item)  # e.g. sorted letters for anagram grouping
    groups.setdefault(key, []).append(item)
```

> **Note on `setdefault`:** `d.setdefault(key, [])` returns `d[key]` if it exists, otherwise inserts `[]` and returns it. It's cleaner than `if key not in d: d[key] = []`.

## Problem 3 — Longest Consecutive Sequence (LC #128) ⭐⭐ Medium

Find the length of the longest sequence of consecutive integers. Must run in O(n).

```
[100, 4, 200, 1, 3, 2]       →  4   (sequence 1,2,3,4)
[0, 3, 7, 2, 5, 8, 4, 6, 0, 1]  →  9
```

**The O(n log n) approach is obvious:** sort, then scan. But sorting costs O(n log n) and the problem demands O(n).

**The O(n) insight requires a hash set:**

If you put all numbers in a set, checking whether `n+1` exists is O(1). But you'd still be re-counting the same sequences. The trick: only start counting from the **beginning** of a sequence. A number `n` is the start of a sequence if and only if `n-1` is NOT in the set.

So: iterate through numbers. If `num - 1` is not in the set, `num` is a sequence start — count how long the sequence is. If `num - 1` IS in the set, skip it (you'll count this sequence from its start).

Why is this still O(n) even though you count from sequence starts? Each number is visited at most twice — once in the outer loop, at most once in the inner counting loop.

````python
# ============================================================
# SOLUTION
# ============================================================

def longestConsecutive_solution(nums: List[int]) -> int:
    num_set = set(nums)
    best = 0

    for num in num_set:
        if num - 1 not in num_set:      # num is the start of a sequence
            length = 1
            while num + length in num_set:
                length += 1
            best = max(best, length)

    return best
````
