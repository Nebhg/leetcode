# Pattern: Dynamic Programming

## What is DP — the idea, no code

DP applies when a problem has two properties:

1. **Overlapping subproblems** — you'd end up solving the same smaller problem multiple times in a naive recursive approach
2. **Optimal substructure** — the best answer to the full problem can be built from the best answers to its subproblems

The solution: compute each subproblem once, store it, look it up instead of recomputing. This turns exponential recursion into linear work.

For every DP problem, work through these in order:

3. **What are the base cases?** What are the smallest inputs you can answer directly, without the recurrence?
4. **What order do you fill the table?** Left to right? Right to left? Does it matter?
5. **What is the final answer?** Is it `dp[n]`? `max(dp)`? `dp[n][m]`? Something else?

You're robbing houses on a street. Each house has some amount of money. You **cannot rob two adjacent houses** (the alarm goes off). Maximise the total money stolen.

nums = [1,2,3,1]    →  4   (rob house 0 and house 2: 1+3)

````python
from typing import List

def rob(nums: List[int]) -> int:
    pass

assert rob([1,2,3,1]) == 4
assert rob([2,7,9,3,1]) == 12
assert rob([1]) == 1
assert rob([2,1]) == 2
print('Problem 2: all passed')
````

<details>
<summary>💡 Hint — what are your two options at house i? (click to expand)</summary>

At house `i`, you either skip it (in which case the best you've done up to `i` is the same as the best up to `i-1`), or you rob it (in which case you get `nums[i]` plus the best you could do two houses back, since you can't have robbed `i-1`). 

So: `dp[i] = max(dp[i-1], nums[i] + dp[i-2])`.

Base case for `dp[1]`: you can only pick one of the first two houses, so `dp[1] = max(nums[0], nums[1])`.
</details>

````python
# ============================================================
# SOLUTION — only run after a genuine attempt
# ============================================================

def rob_solution(nums: List[int]) -> int:
    if len(nums) == 1:
        return nums[0]

    # dp[i] = max money robbing houses 0..i
    # dp[i] = max(dp[i-1], nums[i] + dp[i-2])
    dp = [0] * len(nums)
    dp[0] = nums[0]
    dp[1] = max(nums[0], nums[1])

    for i in range(2, len(nums)):
        dp[i] = max(dp[i - 1], nums[i] + dp[i - 2])

    return dp[-1]

assert rob_solution([1,2,3,1]) == 4
assert rob_solution([2,7,9,3,1]) == 12
print('Solution verified')
````

<details>
<summary>💡 Hint 1 — how to initialise? (click to expand)</summary>

You want to take a `min()` over possible coin choices. If you initialise `dp[i]` to 0, the min will always pick 0 and give wrong answers. Initialise to something that can never be a valid answer — `float('inf')` works well. Then `dp[0] = 0` (zero coins to make zero).

<summary>💡 Hint 2 — stuck on the recurrence? (click to expand)</summary>

```python

    for c in coins:                  # try each coin
        if c <= i:                   # only if the coin fits

```

At the end: `return dp[amount] if dp[amount] != float('inf') else -1`
</details>

````python
# ============================================================
# SOLUTION — only run after a genuine attempt
# ============================================================

def coinChange_solution(coins: List[int], amount: int) -> int:
    # dp[i] = minimum coins to make amount i
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0  # 0 coins to make amount 0

    for i in range(1, amount + 1):
        for c in coins:
            if c <= i:
                dp[i] = min(dp[i], dp[i - c] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1

assert coinChange_solution([1,2,5], 11) == 3
assert coinChange_solution([2], 3) == -1
assert coinChange_solution([1,5,11], 15) == 3
print('Solution verified')
````

<details>
<summary>💡 Hint — why isn't the answer dp[n-1]? (click to expand)</summary>

`dp[i]` is the length of the longest increasing subsequence that **ends at index i**. The overall longest subsequence can end at any index — not necessarily the last one. For example, in `[10,9,2,5,3,7,101,18]`, the LIS ends at index 6 (value 101), not the last index.

So the answer is `max(dp)`, not `dp[-1]`.

The recurrence:
```python
for i in range(1, n):
    for j in range(i):
        if nums[j] < nums[i]:
            dp[i] = max(dp[i], dp[j] + 1)
```
</details>

````python
# ============================================================
# SOLUTION — only run after a genuine attempt
# ============================================================

def lengthOfLIS_solution(nums: List[int]) -> int:
    n = len(nums)
    # dp[i] = length of longest increasing subsequence ending at index i
    dp = [1] * n  # every element is a subsequence of length 1

    for i in range(1, n):
        for j in range(i):
            if nums[j] < nums[i]:          # can extend subsequence ending at j
                dp[i] = max(dp[i], dp[j] + 1)

    return max(dp)  # LIS can end at any index

assert lengthOfLIS_solution([10,9,2,5,3,7,101,18]) == 4
assert lengthOfLIS_solution([0,1,0,3,2,3]) == 4
assert lengthOfLIS_solution([7,7,7,7]) == 1
print('Solution verified')
````
