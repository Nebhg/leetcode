# Pattern: Binary Search

## The idea — no code yet

Binary search works on **sorted** data. You have a search space and you want to find something in it. Instead of checking every element, you check the middle — and based on what you find, you throw away *half* the remaining search space. Repeat until done.

**When to think binary search:**
- The array is sorted (or you can define a sorted search space)
- You're looking for a specific value, or a boundary (first/last position where something is true)
- A linear scan would work but the problem implies or requires O(log n)

**Complexity:** O(log n) time, O(1) space.

</details>

````python
# ============================================================
# SOLUTION — only run this after a genuine attempt above
# ============================================================

def search_solution(nums: List[int], target: int) -> int:
    left, right = 0, len(nums) - 1

    while left <= right:                        # <= so single-element arrays work
        mid = left + (right - left) // 2        # avoids overflow (good habit)

        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1                      # target in right half
        else:
            right = mid - 1                     # target in left half

    return -1

assert search_solution([-1, 0, 3, 5, 9, 12], 9) == 4
assert search_solution([-1, 0, 3, 5, 9, 12], 2) == -1
print('Solution verified')
````

<details>
<summary>💡 Hint — what does left point to after the loop? (click to expand)</summary>

When the loop ends without finding the target, `left` has moved past `right`. At that point, `left` is pointing to the position where the target *would* be inserted to maintain sorted order. No extra logic needed — just return `left`.
</details>

````python
# ============================================================
# SOLUTION — only run this after a genuine attempt above
# ============================================================

def searchInsert_solution(nums: List[int], target: int) -> int:
    left, right = 0, len(nums) - 1

    while left <= right:
        mid = left + (right - left) // 2
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return left  # left is now the insertion point

assert searchInsert_solution([1,3,5,6], 2) == 1
assert searchInsert_solution([1,3,5,6], 7) == 4
print('Solution verified')
````

<details>
<summary>💡 Hint — which half to discard? (click to expand)</summary>

If `nums[mid] > nums[right]`: the left portion up to mid is sorted and the rotation point (where the minimum lives) is to the right of mid. So move left to `mid + 1`.

If `nums[mid] < nums[right]`: the right portion is sorted and the minimum is somewhere in the left half — possibly mid itself. So move right to `mid` (not `mid - 1`, because mid could be the answer).

Key: use `right = mid` not `right = mid - 1` here, because you don't want to exclude the potential minimum.
</details>

````python
# ============================================================
# SOLUTION — only run this after a genuine attempt above
# ============================================================

def findMin_solution(nums: List[int]) -> int:
    left, right = 0, len(nums) - 1

    while left < right:                        # boundary search: stops when left == right
        mid = left + (right - left) // 2

        if nums[mid] > nums[right]:            # rotation point is in the right half
            left = mid + 1
        else:                                  # minimum is in the left half (mid could be it)
            right = mid

    return nums[left]                          # left == right, both point at minimum

assert findMin_solution([3,4,5,1,2]) == 1
assert findMin_solution([4,5,6,7,0,1,2]) == 0
assert findMin_solution([11,13,15,17]) == 11
print('Solution verified')
````
