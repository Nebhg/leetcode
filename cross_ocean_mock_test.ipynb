Return the largest number in the list. `numbers` has at least one element, integers only (can be negative).

### 2c — `second_largest(numbers)`
Return the second-largest *distinct* value. `[5, 1, 4, 2, 4]` → `4` (not `5` again — duplicates of the max don't count as a second value). Assume at least two distinct values exist.

```python
# YOUR TURN
# sort the list and then take the second second to last item 
from typing import List
from numpy import sort


def second_largest(numbers: List[int]) -> int: 
    sorted = sort(numbers)
    # return second to last item in the sorted list
    # remove duplicates by converting to a set and back to a list
    sorted = list(set(sorted))
    return sorted[-2]

assert second_largest([5, 1, 4, 2, 4]) == 4
assert second_largest([10, 10, 9]) == 9
assert second_largest([3, 1]) == 1
print("2c passed")
```

```python
# YOUR TURN
def calculate_total_price(prices: List[int], discount: int) -> int:
    total = sum(prices)
    discounted_total = total * (1 - discount / 100)
    return int(discounted_total)

assert calculate_total_price([10, 20, 30], 0) == 60
assert calculate_total_price([100, 200], 50) == 200
assert calculate_total_price([19.99, 5.50, 12.25], 10) == 35
print("2e passed")
```
